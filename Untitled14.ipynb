1. Perform basic color-based segmentation to separate the blue color in an image?
Convert the image to the HSV color space.
Define HSV ranges for blue.
Use cv2.inRange() to create a mask.
Apply the mask to extract the blue regions.


```python
import cv2
import numpy as np

image = cv2.imread('download (1).jpeg')
hsv_image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

# Define blue color range
lower_blue = np.array([100, 150, 50])
upper_blue = np.array([140, 255, 255])

# Create a mask
mask = cv2.inRange(hsv_image, lower_blue, upper_blue)
segmented_image = cv2.bitwise_and(image, image, mask=mask)

cv2.imshow("Original Image", image)
cv2.imshow("Blue Segmented", segmented_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

3. Load a pretrained Mask R-CNN model from PyTorch and use it for object detection and segmentation on an image? 
Load the torchvision pretrained Mask R-CNN model.
Pass the image through the model.
Display the detected objects and masks.

```python
import torch
from torchvision import models, transforms
from PIL import Image
import matplotlib.pyplot as plt

# Load Mask R-CNN model
model = models.detection.maskrcnn_resnet50_fpn(pretrained=True)
model.eval()

# Preprocess image
image = Image.open('download (1).jpeg').convert("RGB")
transform = transforms.Compose([
    transforms.ToTensor(),
])
image_tensor = transform(image).unsqueeze(0)

# Perform detection
outputs = model(image_tensor)[0]

# Display bounding boxes and masks
plt.imshow(image)
for box, score in zip(outputs['boxes'], outputs['scores']):
    if score > 0.5:
        plt.gca().add_patch(
            plt.Rectangle((box[0], box[1]), box[2]-box[0], box[3]-box[1], linewidth=2, edgecolor='red', facecolor='none')
        )
plt.show()
    
```

Extract the boxes from the model output and draw them on the image using OpenCV.

7. Apply Mask R-CNN to detect objects and their segmentation masks in a custom image and display them?
Follow the steps for loading a pretrained Mask R-CNN and customize the preprocessing pipeline for your image. Display results using Matplotlib.

```python
import torch
import torchvision
from torchvision.transforms import functional as F
from PIL import Image
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import numpy as np

# Load pre-trained Mask R-CNN model
model = torchvision.models.detection.maskrcnn_resnet50_fpn(pretrained=True)
model.eval()  # Set the model to evaluation mode

# Load the custom image
image_path = "download (1).jpeg"  # Replace with your image path
image = Image.open(image_path).convert("RGB")

# Transform the image to a tensor
transformed_image = F.to_tensor(image)  # Convert image to a PyTorch tensor
# Perform inference
with torch.no_grad():
    predictions = model([transformed_image])
# Extract bounding boxes, masks, labels, and scores
pred_boxes = predictions[0]['boxes'].numpy()
pred_masks = predictions[0]['masks'].squeeze(1).numpy()  # Remove extra dimensions
pred_labels = predictions[0]['labels'].numpy()
pred_scores = predictions[0]['scores'].numpy()

# Filter by confidence threshold
threshold = 0.5
filtered_indices = pred_scores > threshold
filtered_boxes = pred_boxes[filtered_indices]
filtered_masks = pred_masks[filtered_indices]
filtered_labels = pred_labels[filtered_indices]
filtered_scores = pred_scores[filtered_indices]

def apply_mask(image, mask, color):
    """Apply a segmentation mask to the image with a given color."""
    for c in range(3):  # Apply the mask to all color channels
        image[:, :, c] = np.where(mask, color[c] * 255, image[:, :, c])
    return image

# Convert image to NumPy array
image_np = np.array(image)

# Generate random colors for each object
colors = np.random.uniform(0, 1, size=(len(filtered_masks), 3))

# Apply each mask to the image
for i, mask in enumerate(filtered_masks):
    mask = mask > 0.5  # Threshold to create a binary mask
    image_np = apply_mask(image_np, mask, colors[i])

# Plot the image with bounding boxes
fig, ax = plt.subplots(1, figsize=(12, 9))
ax.imshow(image_np)

# Draw bounding boxes
for i, box in enumerate(filtered_boxes):
    x1, y1, x2, y2 = box
    rect = patches.Rectangle((x1, y1), x2 - x1, y2 - y1, linewidth=2, edgecolor='red', facecolor='none')
    ax.add_patch(rect)
    ax.text(x1, y1 - 10, f"Class: {filtered_labels[i]} | Score: {filtered_scores[i]:.2f}", color='white', fontsize=10, bbox=dict(facecolor='red', alpha=0.5))

plt.axis('off')
plt.show()

```

8. Apply k-means clustering for segmenting regions in an image
Reshape the image to a 2D array.
Apply k-means clustering using cv2.kmeans().
Reshape the segmented output back to the original dimensions.

```python
image = cv2.imread('download (1).jpeg')
pixel_values = image.reshape((-1, 3))
pixel_values = np.float32(pixel_values)

# Define criteria and number of clusters (k)
criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 100, 0.2)
k = 3
_, labels, centers = cv2.kmeans(pixel_values, k, None, criteria, 10, cv2.KMEANS_RANDOM_CENTERS)

# Convert back to uint8 and reshape to image dimensions
centers = np.uint8(centers)
segmented_image = centers[labels.flatten()]
segmented_image = segmented_image.reshape(image.shape)

cv2.imshow("Segmented Image", segmented_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
